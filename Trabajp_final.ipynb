```python
import pandas as pd
import numpy as np
from os import read
df = pd.read_excel("/content/Base_de_Datos_Medicamentos.xlsx")
```

```python
#1. Filtrar datos
Precio_inf_40= df[df["Precio (USD)"] < 40]
print(Precio_inf_40)
#Segun los datos obtenidos un solo un 39% de los medicamentos tiene un precio inferior de 40 USD

```

```python
#4. Ordenar datos
df_sorted = df.sort_values(by= "ID", ascending =False )
print(df_sorted)
```

```python
#5. Recuento de datos
print (df["Forma Farmacéutica"].value_counts(),df["Laboratorio"].value_counts())
```

```python
#6. Tabla pivot para saber cuantos medicamentos por forma farmaceutica tiene cada laboratorio
pivot_table = df.pivot_table(index="Laboratorio",
                            columns="Forma Farmacéutica",
                            aggfunc="size",
                            fill_value=0)
print(pivot_table)

print("Pfizer es el laboratorio con mas presencia en este listado, siendo el que la mayoria de veces tiene mas medicamentos por forma farmaceutica y siendo el que mas aparece en el listado")
```

```python
#7. Precio acumulado
df[ "Cumulative_Precio"]= df["Precio (USD)"].cumsum()

print (df["Cumulative_Precio"])
```

```python
# NUMPY
df = pd.read_excel("/content/Base_de_Datos_Medicamentos.xlsx")
#1. CALCULOS ESTADISTICOS
media_precio = np.mean(df["Precio (USD)"])
mediana_precio = np.median(df["Precio (USD)"])
desviacion_precio = np.std(df["Precio (USD)"])
print(f"Media del Precio: {media_precio}")
print(f"Mediana del Precio: {mediana_precio}")
print(f"Desviación Estándar del Precio: {desviacion_precio}")
```

```python
#2. CLASIFICACION DE MEDICAMENTOS SEGUN EL PRECIO
df["Categoría de Precio"] = np.where(df["Precio (USD)"] > media_precio, "Costoso", "Económico")
print(df["Categoría de Precio"])
print(df["Categoría de Precio"].value_counts())
```

```python
#3. PERCENTILES
percentil_25 = np.percentile(df["Precio (USD)"], 25)
percentil_50 = np.percentile(df["Precio (USD)"], 50)  # Equivalente a la mediana
percentil_75 = np.percentile(df["Precio (USD)"], 75)
print(f"Percentil 25: {percentil_25}")
print(f"Percentil 50 (Mediana): {percentil_50}")
print(f"Percentil 75: {percentil_75}")
```

```python
#4. Transformacion logaritmica
df["Precio Log"] = np.log1p(df["Precio (USD)"])
print(df["Precio Log"])
```

```python
#5. Correlación entre Variables Clave
correlacion_precio_demanda = np.corrcoef(df["Precio (USD)"], df["Nivel de Demanda (%)"])[0, 1]
print(f"Correlación entre Precio y Nivel de Demanda: {correlacion_precio_demanda}")
```

```python
#GRAFICOS MATPLOTLIB Y SEABORN
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
from mpl_toolkits.mplot3d import Axes3D
```

```python
#GRAFICO DE BARRAS
# Calcular la cantidad de medicamentos por laboratorio
medicamentos_por_laboratorio = df['Laboratorio'].value_counts()

# Crear el gráfico de barras
plt.figure(figsize=(10, 6))  # Ajustar el tamaño del gráfico si es necesario
plt.bar(medicamentos_por_laboratorio.index, medicamentos_por_laboratorio.values, color='skyblue')
plt.xlabel('Laboratorio')
plt.ylabel('Cantidad de Medicamentos')
plt.title('Cantidad de Medicamentos por Laboratorio')
plt.xticks(rotation=45, ha='right')  # Rotar las etiquetas del eje x para mejor visualización
plt.tight_layout()  # Ajustar el diseño para evitar que las etiquetas se superpongan
plt.show()

print("En este grafico tambien se puede observar como pfizer es el laboratorio lider en el mercado, seguido de otros como Sanofi, J&J, Roche y Bayer")
```

```python
#2. GRAFICO DE TORTA
# Calcular la cantidad de cada forma farmacéutica
forma_farmaceutica_counts = df['Forma Farmacéutica'].value_counts()

# Crear el gráfico de torta
plt.figure(figsize=(8, 8))  # Ajusta el tamaño si es necesario
plt.pie(forma_farmaceutica_counts, labels=forma_farmaceutica_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Distribución de Formas Farmacéuticas')
plt.axis('equal')  # Asegura que el gráfico sea un círculo
plt.show()


print("La forma farmaceutica con mas presencia en el mercado es el jarabe, esto puede deberse a sus ventajas como la facilidad en la ingesta y el sabor; lo que lo hace adecuado para niños y adultos mayores")
```

```python
#3. MAPA DE CALOR
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

# Crear la tabla pivote
pivot_table = df.pivot_table(index="Laboratorio",
                            columns="Forma Farmacéutica",
                            aggfunc="size",
                            fill_value=0)
# Crear el mapa de calor
plt.figure(figsize=(10, 8))  # Ajustar el tamaño de la figura
sns.heatmap(pivot_table, cmap="hot", annot=True, fmt="d")
plt.title("Mapa de Calor de Medicamentos por Laboratorio y Forma Farmacéutica")
plt.xlabel("Forma Farmacéutica")
plt.ylabel("Laboratorio")
plt.show()

print("En este mapa de calor se puede observar cuantos medicamentos por forma farmaceutica tiene cada empresa.")
print("Se observa que cada empresa gana de cierta forma en cada forma farmaceutica. J&J tiene mas unguentos, Roche y")
print("sanofi tienen mas tabletas y pfizer tiene mas capsulas y Jarabes")
```

```python
#4. GRAFICO DE DISPERSIÓN 3D
from mpl_toolkits.mplot3d import Axes3D

# Asegurarse de que las columnas sean numéricas
df["Precio (USD)"] = pd.to_numeric(df["Precio (USD)"], errors='coerce')
df["Nivel de Demanda (%)"] = pd.to_numeric(df["Nivel de Demanda (%)"], errors='coerce')
df["ID"] = pd.to_numeric(df["ID"], errors='coerce')


# Crear la figura 3D
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

# Datos para el gráfico
x = df["ID"]
y = df["Precio (USD)"]
z = df["Nivel de Demanda (%)"]


# Crear el gráfico de dispersión
ax.scatter(x, y, z, c='orange', marker='o')

# Etiquetas y título
ax.set_xlabel('Cantidad vendida')
ax.set_ylabel('Precio (USD)')
ax.set_zlabel('Nivel de Demanda (%)')
ax.set_title('Gráfico de Dispersión 3D')

# Mostrar el gráfico
plt.show()

print("Algunos medicamentos con precio bajo tienen una alta cantidad vendida, lo que podría indicar que son de uso frecuente y accesibles para los consumidores.")
```

```python
#5. GRAFICO DE LINEAS Y ALGUNOS RECURSO DE LA LIBRERIA PANDAS
# 1. Verificación de las columnas
print("Columnas del DataFrame:")
print(df.columns)

# 2. Verificación de los tipos de datos
print("\nTipos de datos del DataFrame:")
print(df.dtypes)

# 3. Conversión a numérico y manejo de errores (en caso de que no sean numéricos)
df["Nivel de Demanda (%)"] = pd.to_numeric(df["Nivel de Demanda (%)"], errors='coerce')
df["Precio (USD)"] = pd.to_numeric(df["Precio (USD)"], errors='coerce')

# 4. Verificación de valores NaN
print("\nValores NaN en 'Nivel de Demanda (%)' y 'Precio (USD)':")
print(df[['Nivel de Demanda (%)', 'Precio (USD)']].isnull().sum())

# 5. Eliminar filas con valores NaN
df_clean = df.dropna(subset=["Nivel de Demanda (%)", "Precio (USD)"])

# 6. Ordenar el DataFrame por 'Nivel de Demanda (%)'
df_ordenado = df_clean.sort_values(by='Nivel de Demanda (%)', ascending=True)

# 7. Verificación de la ordenación
print("\nPrimeras filas del DataFrame ordenado:")
print(df_ordenado[['Nivel de Demanda (%)', 'Precio (USD)']].head())


# 8. Crear el gráfico de líneas ordenado
plt.plot(df_ordenado["Nivel de Demanda (%)"], df_ordenado["Precio (USD)"])
plt.title("Gráfico de Línea - Precio de Medicamentos por Nivel de Demanda (Ordenado)")
plt.xlabel("Nivel de Demanda (%)")
plt.ylabel("Precio (USD)")
plt.show()

print("Se puede decir que el precio de los medicamentos no tienen una relacion lineal con respecto al nivel de demanda, en este caso no podemos concluir que el nivel de demanda afecte al precio.")
```
